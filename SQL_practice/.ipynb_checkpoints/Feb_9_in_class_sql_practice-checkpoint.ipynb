```python
import sqlite3
import pandas as pd
```

```python
airports_query = """
SELECT *
FROM airports
"""
```

```python
with sqlite3.connect('data/flights.db') as con:
    airports_df = pd.read_sql(airports_query, con)
```

```python
airports_df
```

```python
schema_query = """
SELECT *
FROM sqlite_master
"""
```

```python
with sqlite3.connect('data/flights.db') as con:
    schema_df = pd.read_sql(schema_query, con)
```

```python
schema_df
```

```python
basic_query = """
SELECT name, city, country
FROM airports
"""
```

```python
with sqlite3.connect('data/flights.db') as con:
    basic_df = pd.read_sql(basic_query, con)
```

```python
basic_df.head()
```

```python
distinct_query = """
SELECT DISTINCT country
FROM airports
"""
```

```python
with sqlite3.connect('data/flights.db') as con:
    distinct_df = pd.read_sql(distinct_query, con)
```

```python
distinct_df.head()
```

```python
alias_query = """
SELECT
    name AS "Airport Name",
    country AS "Airport Country"
FROM airports
"""
```

```python
with sqlite3.connect('data/flights.db') as con:
    alias_df = pd.read_sql(alias_query, con)
```

```python
alias_df.head()
```

```python
count_query = """
SELECT COUNT(*) AS "Number of Airports"
FROM airports
"""
```

```python
with sqlite3.connect('data/flights.db') as con:
    count_df = pd.read_sql(count_query, con)
```

```python
count_df
```
